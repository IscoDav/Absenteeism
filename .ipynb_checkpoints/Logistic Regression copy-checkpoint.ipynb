## Standartize the Data

```python
#from sklearn.preprocessing import StandardScaler

#absenteeism_scaler = StandardScaler()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns, copy = True, with_mean = True, with_std = True):
        self.scaler = StandardScaler(copy,with_mean,with_std)
        self.columns = columns 
        self.mean = None
        self.var = None 
    
    def fit(self, X , y = None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X , y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns = self.columns)
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis =1)[init_col_order]
```

```python
unscaled_inputs.columns.values
```

```python
columns_to_scale = ['Month Value',
       'Day of the Week', 'Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index',
       'Children', 'Pets']

# Removed Dummy Variables before scaling in order save their meaning to the dataset

# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
absenteeism_scaler = CustomScaler(columns_to_scale)
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

# Split the Data into Train & Test and Shuffle 

### Import the Relevant Library

```python
from sklearn.model_selection import train_test_split
```

## Split

```python
x_train, x_test, y_train, y_test = train_test_split(unscaled_inputs, targets, test_size = 0.2, random_state = 20)
```

```python
print(x_train.shape, x_test.shape)
```

```python
print(y_train.shape, y_test.shape)
```

```python
x_train
```

## Logistic Regression with Sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics 
```

### Train the Model

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

### Manually Check The Accuracy

```python
model_output = reg.predict(x_train)
```

```python
model_output
```

```python
model_output == y_train
```

```python
np.sum((model_output == y_train)) / model_output.shape[0]
```

## Finding the Intercept and Coefficient 

```python
reg.intercept_
```

```python
reg.coef_
```

```python
feature_name = unscaled_inputs.columns.values
feature_name
```

```python
summary_table = pd.DataFrame(columns =['Feature Name'], data = feature_name)

summary_table['Coefficient'] = np.transpose(reg.coef_)

summary_table
```

```python
summary_table.index  = summary_table.index + 1 
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

## Interpreting the Coefficients 

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
```

```python
 summary_table.sort_values('Odds_ratio', ascending=False)
```
